```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
train_data=pd.read_csv('train.csv')
```

```python
print(train_data.head())
```

```python
train_y=train_data['Item_Outlet_Sales']
train_data=train_data.drop(columns=['Item_Outlet_Sales'],axis=1)
```

```python
train_data=train_data.drop(columns=['Item_Identifier'],axis=1)
```

```python
train_data['Item_Fat_Content'].replace(to_replace =['low fat','LF'], 
                 value ='Low Fat') 
```

```python
train_data['Item_Fat_Content'].replace("reg", 
                 "Regular") 
```

```python
IT=pd.get_dummies(train_data['Item_Type'])
```

```python
OS=pd.get_dummies(train_data['Outlet_Size'])
```

```python
IFC=pd.get_dummies(train_data['Item_Fat_Content'])
```

```python
train_data=train_data.drop(columns=['Item_Type','Outlet_Size','Item_Fat_Content'])
```

```python
result = pd.concat([train_data, IT], axis=1, sort=False)
```

```python
result1 = pd.concat([result, OS], axis=1, sort=False)
```

```python
result2= pd.concat([result1, IFC], axis=1, sort=False)
```

```python
result2.head()
```

```python
OI=pd.get_dummies(result2['Outlet_Identifier'])
```

```python
OLT=pd.get_dummies(result2['Outlet_Location_Type'])
```

```python
OT=pd.get_dummies(result2['Outlet_Type'])
```

```python
rslt=result2.drop(columns=['Outlet_Identifier','Outlet_Location_Type','Outlet_Type'])
```

```python
rslt= pd.concat([rslt, OI], axis=1, sort=False)
```

```python
rslt= pd.concat([rslt, OT], axis=1, sort=False)
```

```python
rslt= pd.concat([rslt, OLT], axis=1, sort=False)
```

```python
train_x=rslt
```

```python
model = LinearRegression()
```

```python
train_x.dropna()
```

```python
model.fit(train_x,train_y)
```

```python
print('\nCoefficient of model :', model.coef_)
```

```python
print('\nIntercept of model',model.intercept_)
```
