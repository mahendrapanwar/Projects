```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
train.Item_Fat_Content=LabelEncoder().fit_transform(train.Item_Fat_Content)
train.Item_Type=LabelEncoder().fit_transform(train.Item_Type)
train.Outlet_Location_Type=LabelEncoder().fit_transform(train.Outlet_Location_Type)
train.Outlet_Type=LabelEncoder().fit_transform(train.Outlet_Type)
train.Outlet_Size=LabelEncoder().fit_transform(train.Outlet_Size)
train.head()
```

```python
model = LinearRegression()
from sklearn.model_selection import train_test_split
train,test = train_test_split(train, test_size=0.33, random_state=42)
test.head()
```

```python
train.dropna(how='any',inplace=True,axis=1)
test.dropna(how='any',inplace=True,axis=1)
trainy=train['Item_Outlet_Sales']
testy=test['Item_Outlet_Sales']
trainx=train.drop(columns=['Item_Outlet_Sales','Outlet_Identifier','Item_Identifier'],axis=1)
testx=test.drop(columns=['Item_Outlet_Sales','Outlet_Identifier','Item_Identifier'],axis=1)
```

```python
mdd=model.fit(trainx,trainy)
```

```python
predict_test = model.predict(testx)
```
