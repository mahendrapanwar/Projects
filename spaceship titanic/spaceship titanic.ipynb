```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
ans=pd.read_csv('test.csv')
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
train.CryoSleep=LabelEncoder().fit_transform(train.CryoSleep)
train.Cabin=LabelEncoder().fit_transform(train.Cabin)
train.Destination=LabelEncoder().fit_transform(train.Destination)
train.Age=LabelEncoder().fit_transform(train.Age)
train.HomePlanet=LabelEncoder().fit_transform(train.HomePlanet)
test.CryoSleep=LabelEncoder().fit_transform(test.CryoSleep)
test.Cabin=LabelEncoder().fit_transform(test.Cabin)
test.Destination=LabelEncoder().fit_transform(test.Destination)
test.Age=LabelEncoder().fit_transform(test.Age)
test.HomePlanet=LabelEncoder().fit_transform(test.HomePlanet)
```

```python
train['sum']=train['RoomService']+train['FoodCourt']+train['ShoppingMall']+train['Spa']+train['VRDeck']
test['sum']=test['RoomService']+test['FoodCourt']+test['ShoppingMall']+test['Spa']+test['VRDeck']
```

```python
train.describe()
```

```python
train['sum']=pd.cut(x=train['sum'],bins=[0,9000,18000,27000,36000],labels=[0,1,2,3])
```

```python
test['sum']=pd.cut(x=test['sum'],bins=[0,9000,18000,27000,36000],labels=[0,1,2,3])
```

```python
train.describe()
```

```python
train['Age']=pd.cut(x=train['Age'],bins[0,15,30,45,60,80],labels=[0,1,2,3,4])
test['Age']=pd.cut(x=test['Age'],bins[0,15,30,45,60,80],labels=[0,1,2,3,4])
```

```python
train.head()
```

```python
train['Transported']=train['Transported'].astype(int)
```

```python
train=train.drop(['PassengerId','Cabin','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Name'],axis=1)
test=test.drop(['PassengerId','Cabin','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Name'],axis=1)
```

```python
train.describe()
```

```python
train.isnull().sum()
```

```python
train.fillna(0, inplace=True)
test.fillna(0, inplace=True)
train.isnull().sum()
```

```python
train['VIP']=train['VIP'].astype(int)
```

```python
test['VIP']=test['VIP'].astype(int)
```

```python
train.head()
```

```python
trainx=train.drop(['Transported'],axis=1)
```

```python
trainy=train['Transported']
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
log_reg=LogisticRegression()
rnd_clf=RandomForestClassifier()
svm_clf=SVC()
dt_clf=DecisionTreeClassifier()
voting_clf=VotingClassifier(estimators=[ ('rf', rnd_clf), ('svc', svm_clf), ('dt', dt_clf)], voting='hard')
```

```python
vot=voting_clf.fit(trainx,trainy)
```

```python
test.head()
```

```python
prediction=vot.predict(test)
```

```python
prediction
```

```python
df = pd.DataFrame(prediction)
```

```python
df.head()
```

```python
df=df.astype(bool)
```

```python
df.head()
```

```python
df.rename(columns={'0':'Transported'},inplace=True)
```

```python
df['PassengerId']=ans['PassengerId']
```
